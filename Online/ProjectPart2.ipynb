Next part is the hard part.  `math` can be used for most constants but `numpy` or `np` will be needed for more complicated parts.  I load and use both packages below.  They are two of my most common to use!

```
import math
import numpy as np

def gaussian(x):
  return 1/math.sqrt(2*math.pi)*np.e**(-x**2/2)

```
