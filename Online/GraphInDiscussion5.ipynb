```python
import graphviz

dot = graphviz.Digraph(comment='ECU Mathematics and Computer Science Courses')
dot.node('DM', 'Discrete Math')
dot.node('P1', 'Programming 1')
dot.node('CA', 'College Algebra')
dot.node('Calc','Calculus and Analytic Geometry')
dot.node('BusCalc','Business Calculus')
dot.node('Trig','Trigonometry')


dot.edge( 'P1','DM')
dot.edge('CA','P1','Co-Requisite')
dot.edge('Calc','DM')
dot.edge('BusCalc','DM')
dot.edge('Calc','BusCalc','Either',constraint='false')
dot.edge('BusCalc','Calc','Or',constraint='false')
dot.edge('CA','Calc')
dot.edge('CA','BusCalc')
dot.edge('Trig','Calc')
dot
```
