Task #1

Objective:

Design a simple Student Information System using Python classes and objects.

The system should be able to store and display information about students without using a constructor.

Requirements:

Create a Student class with the following attributes:

Student ID

Name

Age

Grade

Implement setter methods in the Student class to set the values of the attributes.

Implement a display_info method in the Student class to display information about the student.

Create two instances of the Student class and demonstrate the use of setter methods and the display_info method.

Extend the system by creating a StudentManager class that manages a list of students.

Implement methods in the StudentManager class to:

Add a new student to the list.

Display information about all students in the list.

Create an instance of the StudentManager class and use it to add students and display their information.

```python
#Create a Student Class
class student:

#Give attributes without using __init__ and using setter methods
  def setting_student_id(self, student_id):
    self.student_id = student_id

  def setting_name(self, name):
    self.name = name

  def setting_age(self, age):
    self.age = age

  def setting_grade(self, grade):
    self.grade = grade

#Create a display_info method in the student class to display the information
  def display_info(self):
    print("student ID -- ", self.student_id)
    print("Name -- ", self.name)
    print("Age -- ", self.age)
    print("Grade -- ", self.grade)
    print()

#Create two instances of the Student class using setter method and make sure they display
#student one
student1 = student()
student1.setting_student_id(1)
student1.setting_name("Casey Mayo")
student1.setting_age(32)
student1.setting_grade("Senior")
#student two
student2 = student()
student2.setting_student_id(2)
student2.setting_name("Aby Mayo")
student2.setting_age(14)
student2.setting_grade("Freshman")

#Make sure both student's information displays correctly.
print("Student ID 1:")
student1.display_info()

print("Student ID 2:")
student2.display_info()

#Extend the system by creating a StudentManager class that manages a list of students.
class StudentManager:
  students = []

  #Giving attributes to the StudentManager class.
  #Adding a new student
  def add_new_student(self, student):
    self.students.append(student)

  #Displaying all the students
  def display_all_students(self):
    if not self.students: #checking to make sure we have a list
      print("No students to list.")
    else:
      for idx, student in enumerate(self.students, start=1): #creating an index of student to display info in a loop.
        print(f"student's {idx} number:")                    #Starting at 1, because that's what the ID number is we started with.
        student.display_info()
        print() #space between prints.


#Create an instance of the StudentManager class and use it to add students and display their information.
Manager = StudentManager()

#Making sure our previous student's are on the Manager's list
Manager.add_new_student(student1)
Manager.add_new_student(student2)

#Function to add a new student to the list
def add_new_student(Manager):
  new_student = student()
  new_student.setting_student_id(input("Enter Student ID Number: "))
  new_student.setting_name(input("Enter the Student's full name: "))
  new_student.setting_age(input("Enter the Student's age: "))
  new_student.setting_grade(input("Enter the Student's grade: "))
  Manager.add_new_student(new_student)
  print("Completed.")
  print()

#Testing the addition of new students
add_new_student(Manager)

Manager.display_all_students()
```

Objective:

Design a simple Library Management System using Python classes and objects. The system should be able to store and display information about books and library patrons without using a constructor.

Requirements:

Create a Book class with the following attributes:

Book ID

Title

Author

Genre

Availability (True if available, False if checked out)

Implement setter methods in the Book class to set the values of the attributes.

Implement a display_info method in the Book class to display information about the book.

Create two instances of the Book class and demonstrate the use of setter methods and the display_info method.

Create a Patron class with the following attributes:

Patron ID

Name

Checked Out Books (a list of books checked out by the patron)

Implement setter methods in the Patron class to set the values of the attributes.

Implement a display_info method in the Patron class to display information about the patron.

Create two instances of the Patron class and demonstrate the use of setter methods and the display_info method.

```python
#Create a Book Class
class book:

#Give attributes without using __init__ and using setter methods
#Just took the class and attributes from the previous assignment and changed names.
  def setting_book_id(self, book_id):
    self.book_id = book_id

  def setting_title(self, title):
    self.title = title

  def setting_author(self, author):
    self.author = author

  def setting_genre(self, genre):
    self.genre = genre
#Adding availability attribute.
  def setting_availability(self, availability):
    self.availability = availability

#Create a display_info method in the student class to display the information
  def display_info(self):
    print("Book ID -- ", self.book_id)
    print("Title -- ", self.title)
    print("Author -- ", self.author)
    print("Genre -- ", self.genre)
    #Use if/else to check availability
    print("Availability -- ", "Available" if self.availability else "Checked Out")
    print()

#Create two book instances using setter method and then display them.
book1 = book()
book1.setting_book_id(1)
book1.setting_title("Look Alive Twenty Five")
book1.setting_author("Janet Evanovich")
book1.setting_genre("Fiction")
book1.setting_availability(False)

book2 = book()
book2.setting_book_id(2)
book2.setting_title("The 5AM Club")
book2.setting_author("Robin Sharma")
book2.setting_genre("Self Help")
book2.setting_availability(False)

#Testing out displaying our books
print("Testing book 1")
print()
book1.display_info()

print("Testing book 2")
print()
book2.display_info()

#Creating a Patron class
class patron:

  def setting_patron_id(self, patron_id):
    self.patron_id = patron_id

  def setting_name(self, name):
    self.name = name

  def setting_checked_books(self, checked_books):
    self.checked_books = checked_books

#Create a display_info method in the patron class to display the information
  def display_info(self):
    print("Patron ID : ", self.patron_id)
    print("Name : ", self.name)
    print("Checked Out Books : ", self.checked_books)
    print()

#Create two patron instances using setter method and then display them.
patron1 = patron()
patron1.setting_patron_id(1)
patron1.setting_name("Casey Mayo")
patron1.setting_checked_books(book2.title)

patron2 = patron()
patron2.setting_patron_id(2)
patron2.setting_name("Aby Mayo")
patron2.setting_checked_books(book1.title)

#Display the info.
print("Patron 1 Info:  ")
patron1.display_info()

print("Patron 2 Info:  ")
patron2.display_info()
```
