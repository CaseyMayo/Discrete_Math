Below, I am adding Dr. Jacob's python code definition of adding one number to a binary number.

```python
def addone(list7binary):
  counter = 7
  list7binary[counter] = str(1 + int(list7binary[counter]))
  while list7binary[counter]=='2':
    list7binary[counter] = '0'
    counter -= 1
    list7binary[counter] = str(1 + int(list7binary[counter]))
  return list7binary

addone(['0','0','1','0','0','1','0','1'])
```

Below, I am adding Dr. Jacob's python code definition of flipping the numbers. 

```python
def flipit(list7binary):
  newlist = []
  for i in list7binary:
    if i == '0':
      newlist.append('1')
    else:
      newlist.append('0')
  return newlist

flipit(['0','0','1','0','0','1','1','0'])
```

Below, is Dr. Jacob's code putting the above definitions together and labeling it's 'twocomplement'.

```python
def twoscomplement(x):
  if not (x>-128)&(x<127):
    return "No Two's Complement"
  ans = '00000000'
  ans = list(ans) #strings don't allow you to change them in place (immutable) but lists can be!
  counter = 7
  if x>=0:
    while x != 0:
      ans[counter] = str(x%2)
      x = floor(x/2)
      counter -= 1
  if x<0:
    x = abs(x)
    while x != 0:
      ans[counter] = str(x%2)
      x = floor(x/2)
      counter -= 1
    ans = flipit(ans)
    addone(ans)        
  ans = ''.join(ans)
  return ans
```
