# 2 Creating a Universal Set

Just kidding on the maths thing. We're going to try some Python coding. 

First thing first, let's create a universal set. 

Because Universal is a long word.. let's call our Universal Set $Music$.



````
SetofSets = [SchoolBand,Orchestra,GarageBand]
Music = set()
for sett in SetofSets:
  Music = Music.union(sett)       #In your example project, did you use "sett" because "set" is a python dictionary word? 
````


# 3 But **WAIT** there's more!

We need add some instruments that weren't included in the original sets.

````
LetsGetElectric = {'Guitar', 'Turn Table', 'Synthesizer', 'Sampler', 'Keyboard'}

Music = Music.union(LetsGetElectric)
````

# 7 Power sets? or did you mean... **POWER BUTTON**

````
def DPpowerset(s):                                       #Using this as definition for when I call the powerset of each list.
  powerset = [[]]
  for i in s:
    for sett in powerset:
      powerset = powerset + [sett + [i]]
  return powerset
````

The **POWERSET** of the $Music$ (Universal set) would contain a whopping 262,144 elements!

````
DPpowerset(SchoolBand)
````
