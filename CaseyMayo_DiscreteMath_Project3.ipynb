A ***Compound Event*** is the probability of more than one event occuring together. 

Example: Drawing a ***club***, then drawing a second time and the card also being a ***Club*** (No replacement).

P(1st Club) = 13/52

P(2nd Club) = 12/52 (we didn't replace the first card before drawing again)

```
p =[(13/52)*(12/52)]

sum(p)
```

P(Question)={P(Spade)+P(Ace)-P(SpadeAndAce)}

***Computing Conditional Probabilities***

***With Bayes Theorem for verification and explanation***

***Union*** example... What is the conditional probability that The random card drawn was a Spade, if it was known already that the random card drawn was an Ace? 

$$
P(Spade|Ace)=\frac{P(Spade\cap Ace)}{P(Ace)}
$$

***Apply Bayes Theorem***

$$
P(A|B) = \frac{P(A)P(B|A)}{P(B)}
$$

```
a = {(1/52) / (4/52)} #Finding P(Spade|Ace)

sum(a)
```

***SIMULATION TIME***

```
from random import choices


choices(['Spades','Clubs','Diamonds','Hearts'],k=10)
```

Sampling an event 10 times doesn't give a wide enough picture. Maybe.... because I have four possibilities and am only gathering 10 samples. They're bound to be skewed. 

What if we did 10000 samples?

```
HundoRolls =  choices(['Spades','Clubs','Diamonds','Hearts'],k=10000)

countSpades = 0
countClubs = 0
countDiamonds = 0
countHearts = 0

for i in HundoRolls:
  if i =='Spades':
    countSpades += 1
  elif i == 'Clubs':
    countClubs += 1
  elif i == 'Diamonds':
    countDiamonds += 1   
  elif i == 'Hearts':
    countHearts += 1

estimateProbs = [countSpades/len(HundoRolls),countClubs/len(HundoRolls),countDiamonds/len(HundoRolls),countHearts/len(HundoRolls)]
```

```
estimateProbs
```
