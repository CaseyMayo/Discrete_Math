$ \lnot (r\oplus p) \Rightarrow (p \lor q)$

```
values = [True,False]
table = []

for r in values:
  for p in values:
    for q in values: 
      row =[]
      row.append(r)
      row.append(p)
      row.append(q)
      row.append(not(r))
      row.append(not(r) and p)
      row.append(q and p)
      row.append(implies(not(r) and p,q and p))
      table.append(row)


```
