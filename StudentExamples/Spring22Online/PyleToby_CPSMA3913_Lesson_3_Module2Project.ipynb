Let's say I've just ordered a pound o' dice off of chessex. I open the bag up, I dump the dice out, and I count how many dice there are. I then count how many of each dice there are, given d4, d6, d8, d10, d12, d20, and d100. We won't observe the probability of what a die will roll, but rather which dice I pull out of the bag first.

image.png

Let's say there are 100 dice total in the bag. There 15 d4's, 30 d6's, 15 d8's, 20 d10's, 5 d12's, 10 d20's, and 5 d100's. 15+30+15+20+5+10+5=100. 

Probabilities of pulling each die: d4-3/20, d6-6/20, d8-3/20, d10-4/20,     d12-1/20, d20-2/20, d100-1/20. 

```python
p= [3/20,6/20,3/20,4/20,1/20,2/20,1/20]
sum(p)
```

Here is a compound event where my probability is the likelihood that I will pull either a d6 or a d10.

```python
ProbabilityR = 0

for i in [1,3]:
  ProbabilityR = ProbabilityR + p[i]

ProbabilityR
```

Now, I will show the union and intersection of pulling a d6 or a d10. 

Below is the probability of pulling a d6 and either a d6 or d10 on the first 2 draws. 

```python
intersection = p[1]*ProbabilityR
intersection
```

Here is the probability of pulling a d6 in your second draw given that you got either a d6 or a d10 in your first draw.

```python
RGA = intersection/ProbabilityR
RGA
```

Now, I'll figure the probability of pulling a d6 or a d10 on your second draw given that you got a d6 on the first draw. 

```python
AGR = intersection/p[1]
AGR
```

Next, I'll apply Baye's Theorem.

```python
RGA ==(p[1]* AGR)/ProbabilityR
```

In a simulation, I will find the frequency counts of pulling dice at random. I will multiply the probabilities by 100.

```python
p100 = [100*i for i in p]
p100
```

Now, I will run a simulation that will pull 10 dice at random.

```python
from random import choices


draw = choices(['d4','d6','d8','d10','d12','d20','d100'],p100,k=10)
print(draw)

countd4 = 0;
countd6 = 0;
countd8 = 0;
countd10 = 0;
countd12 = 0;
countd20 = 0;
countd100 = 0;

for i in draw:
  if i == 'd4':
    countd4 += 1
  if i == 'd6':
    countd6 += 1
  if i == 'd8':
    countd8 += 1
  if i == 'd10':
    countd10 += 1
  if i == 'd12':
    countd12 += 1
  if i == 'd20':
    countd20 += 1
  if i == 'd100':
    countd100 += 1

estimateProb = [countd4/len(draw), countd6/len(draw), countd8/len(draw), countd10/len(draw), countd12/len(draw), countd20/len(draw), countd100/len(draw)]
estimateProb
```
