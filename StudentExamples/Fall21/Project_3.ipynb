I will be converting a number 24036 from base 7 to base 10

```python
total = 0
count = 0
number = '24036'
base = 7
for c in map(int, number):
  count += 1
  total += c * (base**(len(number) - count))
total
```

Now I am going to calculate base ten subtraction of numbers 25 and 37 using two's compliment.

The operation I have to do is 25 + (-37). I will first convert these two numbers, 25 and -37 into binary.

```python
from math import floor

count = 0
number1 = 25
number2 = 37
  
def conv(num):
  print(num)
  total = ''
  while num > 1:
    m = num % 2
    num = floor(num/2)
    total = str(m) + total
  total = str(num) + total

  while len(total) < 8:
    total = '0' + total
  
  print(total)

conv(number1)
conv(number2)
```
